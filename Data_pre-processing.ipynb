# Data Understanding

```python
print(f'{'*' * 50} Shape of the Data {'*' * 50}')
display(df.shape)

print(f'{'*' * 50} Information {'*' * 50}')
display(df.info())

print(f'{'*' * 50} Descriptive Statistics {'*' * 50}')
display(df.describe())

# Checking for Outliers 
print(f'{'*' * 50} Outlier Detection {'*' * 50}')
num_col = ['Income','MntWines','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds','NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases']
for col in num_col:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lw = q1 - (1.5*iqr)
    uw = q3 + (1.5*iqr)
    outliers = ((df[col] < lw) | (df[col] > uw))
    print(f'{col} : {round(outliers.sum() / len(df) * 100,2)}')

# Checking for Null Values
print(f'{'*' * 50} Missing Values {'*' * 50}')
round((df.isnull().sum() / len(df) ) * 100,2)
```

# Data Cleaning

```python
def data_cleaning(df):
    # Convert Dt_Customer column into DateTime
    df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format = '%d-%m-%Y')
    
    
    # Fill missing values in Income 
    df = df.fillna({'Income':df['Income'].mean()})
    
    
    # Derived new columns for further analysis 
    # The data is present till 2014 so i calculate age according to that
    df['Age'] = 2014 - df['Year_Birth']
    def age_bins(df):
        if df['Age'] > 0 and df['Age'] <= 30:
            return 'Young'
        elif df['Age'] > 30 and df['Age'] <= 60:
            return 'Middle Aged'
        else:
            return 'Senior'
    df['Age_bins'] = df.apply(age_bins,axis = 1)
    
    
    # Fix the column values
    df['Education'] = df['Education'].replace({'PhD':'Masters','2n Cycle':'Bachelors','Basic':'Undergraduate','Graduation':'Bachelors','Master':'Masters'})
    df['Marital_Status'] = df['Marital_Status'].replace({'Together':'In Couple','Divorced':'Single','Widow':'Single','Alone':'Single','YOLO':'Single','Absurd':'Single','Married':'In Couple'})
    
    
    # Create New Column 
    df['Childrens'] = df['Kidhome'] + df['Teenhome']
    df['Spending'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
    
    
    # According to descriptive summry i created a column for income groups
    def income_groups(df):
        if df['Income'] <= 30000:
            return 'Low Income'
        elif df['Income'] > 30000 and df['Income'] <= 60000:
            return 'Average Income'
        else:
            return 'High Income'
    df['Income_groups'] = df.apply(income_groups,axis = 1)
    
    
    # Rename all the columns for better understanding
    df = df.rename(columns = {'MntWines':'Wines','MntFruits':'Fruits','MntMeatProducts':'Meat',
                        'MntFishProducts':'Fish','MntSweetProducts':'Sweet','MntGoldProds':'Gold',
                        'NumWebPurchases':'Web_Purchases','NumDealsPurchases':'Discount_Purchases','NumCatalogPurchases':'Catelog_Purchases',
                        'NumStorePurchases':'In-Store_Purchases','Response':'Last_Campaign_Response',
                        'AcceptedCmp3':'Accepted_3rd_Campaign', 'AcceptedCmp4':'Accepted_4th_Campaign', 'AcceptedCmp5':'Accepted_5th_Campaign',
                        'AcceptedCmp1':'Accepted_1st_Campaign','AcceptedCmp2':'Accepted_2nd_Campaign'})
    
    
    # Create a new column to see responses out of last 5 campaign 
    df['Average_Response_Rate'] = (df['Accepted_3rd_Campaign'] + df['Accepted_4th_Campaign'] +
                                   df['Accepted_5th_Campaign'] + df['Accepted_1st_Campaign'] +
                                   df['Accepted_2nd_Campaign'])
    
    
    # Drop the unnecessary columns
    df = df[['ID', 'Age', 'Age_bins', 'Education', 'Marital_Status', 'Income', 'Income_groups','Childrens',
            'Dt_Customer', 'Wines', 'Fruits', 'Meat', 'Fish',
           'Sweet', 'Gold', 'Spending', 'Discount_Purchases', 'Web_Purchases',
           'Catelog_Purchases', 'In-Store_Purchases',
           'Accepted_3rd_Campaign', 'Accepted_4th_Campaign',
           'Accepted_5th_Campaign', 'Accepted_1st_Campaign',
           'Accepted_2nd_Campaign', 'Complain',
           'Last_Campaign_Response',
            'Average_Response_Rate']]

    return df
```

```python
cleaned_df = data_cleaning(df)
```

```python
cleaned_df.to_csv('Sale_Conversion_Project_Cleaned.csv')
```
